```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep = '\t')
```

```python
print(chipo.head())
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').str.strip()

chipo['item_price'] = chipo['item_price'].astype(float)

count = chipo[chipo['item_price'] > 10.00]['item_price'].count()

print("Numbers of products cost more than $10.00:", count)
```

```python
products_gt_10 = chipo.loc[chipo['item_price'] > 10.00, 'item_name']

print(products_gt_10)
```

```python
sorted_products = products_gt_10.sort_values()

print(sorted_products)
```

```python
chipo_sorted = chipo.sort_values('item_price', ascending=False)
most_expensive_quantity = chipo_sorted.iloc[0]['quantity']

print("Number of most expensive items ordered:", most_expensive_quantity)
```

```python
veggie_salad_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
veggie_salad_quantity = veggie_salad_orders['quantity'].sum()

print("Number of times Veggie Salad Bowl was ordered:", veggie_salad_quantity)
```

```python
multiple_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
multiple_soda_count = len(multiple_soda_orders)

print("Number of times someone ordered more than one Canned Soda:", multiple_soda_count)
```
