```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
```

```python
tips = pd.read_csv(url)

print(tips.head())
```

```python
tips = tips.drop('Unnamed: 0', axis=1)

print(tips.head())
```

```python
plt.hist(tips['total_bill'], bins=10, edgecolor='black')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10, alpha=0.7)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot: Total Bill, Tip, and Size')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Average Total Bill by Day')
plt.show()
```

```python
plt.scatter(tips['tip'], tips['day'], c=tips['sex'].map({'Male': 'blue', 'Female': 'red'}))
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatter Plot: Tip vs Day (Differentiated by Sex)')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill per Day (Differentiated by Time)')
plt.legend(title='Time', loc='upper right')
plt.show()
```

```python
dinner_tips = tips[tips['time'] == 'Dinner']
lunch_tips = tips[tips['time'] == 'Lunch']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(dinner_tips['tip'], bins=10, edgecolor='black')
axes[0].set_title('Dinner Tips')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')

axes[1].hist(lunch_tips['tip'], bins=10, edgecolor='black')
axes[1].set_title('Lunch Tips')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
male_tips = tips[tips['sex'] == 'Male']
female_tips = tips[tips['sex'] == 'Female']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(male_tips['total_bill'], male_tips['tip'], c=male_tips['smoker'].map({'Yes': 'red', 'No': 'blue'}))
axes[0].set_title('Male Tips')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

axes[1].scatter(female_tips['total_bill'], female_tips['tip'], c=female_tips['smoker'].map({'Yes': 'red', 'No': 'blue'}))
axes[1].set_title('Female Tips')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

legend_labels = ['Non-Smoker', 'Smoker']
colors = ['blue', 'red']
legend_handles = [plt.Line2D([], [], marker='o', color=color, linestyle='', label=label) for color, label in zip(colors, legend_labels)]
plt.legend(handles=legend_handles, loc='upper right')

plt.tight_layout()

plt.show()
```
